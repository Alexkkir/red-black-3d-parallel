```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 4, 8, 16, 32, 64, 128]
y = [4.763589382171631, 2.553377946217855, 1.5416107177734375, 1.0776249567667644, 0.9299062093098959, 1.276364008585612, 2.3686707814534507, 6.975501537322998]
```

```python
from scipy.interpolate import make_interp_spline

X_Y_Spline = make_interp_spline(x, y)
 
# Returns evenly spaced numbers
# over a specified interval.
X_ = np.linspace(min(x), max(x), 500)
Y_ = X_Y_Spline(X_)
```

```python
plt.figure(figsize=(16, 10))
sns.set_style('darkgrid')
sns.lineplot(x, y)
sns.scatterplot(x, y, marker='o')

for a, b in zip(x, y):
    plt.annotate(str(a), (a + 0.01, b + 0.01), fontsize=15)

plt.xlabel(' ', fontsize=20)
plt.xscale('log')
plt.ylabel('  ()', fontsize=20)
plt.title(
    '       MPI', fontsize=25)
plt.savefig('mpi.jpg')
```
